```python
import gradio as gr
from transformers import pipeline
from PIL import Image
import os

os.environ["HF_API_KEY"] = "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"

classifier = pipeline("image-classification", model="google/vit-base-patch16-224")

calorie_database = {
    "apple": 52,
    "banana": 89,
    "orange": 47,
    "pizza": 266,
    "hamburger": 295,
    "french fries": 312,
    "donut": 452,
    "cake": 257,
    "butcher shop, meat market": 60,
    "sushi": 140,
    "ice cream": 207,
}

def classify_food(image):
    predictions = classifier(image)
    food_item = predictions[0]['label']

    calories = calorie_database.get(food_item.lower(), "unknown")

    return food_item, calories


iface = gr.Interface(
    fn=classify_food,
    inputs=gr.Image(type="pil"),
    outputs=[
        gr.Textbox(label="Food Item"),
        gr.Textbox(label="Estimated Calories")
    ],
    title="Food Image Calorie Estimator",
    description="Upload an image of food, and the model will identify the food item and estimate its calorie content."
)

iface.launch()
```
